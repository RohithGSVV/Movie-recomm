```python
import pandas as pd

df = pd.read_csv('Data/Top Ranked Real Movies Dataset.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['Watch Time'] = df['Watch Time'].astype(str).str.replace(' min', '').astype(int)
```

```python
df['Votes'] = df['Votes'].astype(str).str.replace(',', '').astype(int)
```

```python
import re

def extract_year(value):
    match = re.search(r'\d{4}', value)
    if match:
        return int(match.group(0))
    else:
        return None 
df['Year of Release'] = df['Year of Release'].apply(extract_year)
```

```python
df['Gross'] = pd.to_numeric(df['Gross'].str.replace('[$,M]', '', regex=True), errors='coerce')
df = df.rename(columns={'Gross': 'Gross in Mil'})
```
